```python
def split_objects(data):
    en_data = []
    sn_data = []

    for item in data:
        en_data.append({'en': item['en']})
        sn_data.append({'sn': item['sn']})

    return en_data, sn_data
```

```python
def split_to_files(data):
    train_data = data[:20000]
    test_data = data[20000:24000]
    validation_data = data[24000:25000]

    with open('en_sentences_train.txt', 'w', encoding='utf-8') as en_train_file, \
         open('sn_sentences_train.txt', 'w', encoding='utf-8') as sn_train_file, \
         open('en_sentences_test.txt', 'w', encoding='utf-8') as en_test_file, \
         open('sn_sentences_test.txt', 'w', encoding='utf-8') as sn_test_file, \
         open('en_sentences_validation.txt', 'w', encoding='utf-8') as en_validation_file, \
         open('sn_sentences_validation.txt', 'w', encoding='utf-8') as sn_validation_file:
         
        for item in train_data:
            en_train_file.write(item['en'] + '\n')
            sn_train_file.write(item['sn'] + '\n')

        for item in test_data:
            en_test_file.write(item['en'] + '\n')
            sn_test_file.write(item['sn'] + '\n')

        for item in validation_data:
            en_validation_file.write(item['en'] + '\n')
            sn_validation_file.write(item['sn'] + '\n')
```

```python
split_to_files(dataset['train']['translation'])
```
